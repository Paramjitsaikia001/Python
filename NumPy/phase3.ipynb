## working on some real data using numpy


```python

import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# yearly data of some resturant - 2021,2022,2023,2024,2025

# Columns:
# [ID, Year1 Income, Year2 Income, Year3 Income, Year4 Income, Year5 Income]

sales_data = np.array([
    [1, 1250000, 1420000, 1510000, 1700000, 1830000],   # SpiceHub
    [2, 1000000, 1130000, 1200000, 1350000, 1410000],   # FlavourNest
    [3,  820000,  910000,  980000, 1060000, 1140000],   # CurryPoint
    [4, 2050000, 2200000, 2340000, 2480000, 2620000],   # RoyalBite
    [5, 1500000, 1580000, 1690000, 1820000, 1900000],   # UrbanTadka
    [6,  750000,  800000,  880000,  930000, 1000000],   # MealCorner
    [7, 1810000, 1940000, 2020000, 2170000, 2290000],   # GrillStreet
    [8, 1300000, 1390000, 1470000, 1530000, 1610000],   # FoodArc
    [9,  950000, 1020000, 1100000, 1210000, 1290000],   # TastyTreats
    [10,1120000, 1200000, 1340000, 1400000, 1520000]    # ChefVilla
])


print("What is the shape of sales data : ",sales_data.shape)

print("\n what is the income of 1st resto: \n",sales_data[:3,1:])
```

```python
#total sales per year

#NOTE: axios =0 for column and it's by default
print("Total sales per year: \n",np.sum(sales_data[:,1:],axis=0))

#minimum sales per resturant
print("\nMinimum sales per resturant : \n",np.min(sales_data[:,1:],axis=1))

#maximum sales per year
print("\n maximum sales per year: \n",np.max(sales_data[:,1:],axis=0))

#average sales per resturant 
avgperresto=np.mean(sales_data[:,1:],axis=1)
print("\n Average sales per resturant: \n",avgperresto)

plt.figure(figsize=(8,6))
plt.plot(avgperresto)
plt.title("avegare sales per returant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

## Cumulative sum

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)

print("cumulative sum : \n",cumsum)
```
