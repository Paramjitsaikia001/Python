## numpy learning phase 2 is array operatin


```python
import numpy as np
```

```python

arr=np.array([1,2,3,4,5,6,7])

print(arr)
print("basic slicing :",arr[2:5])
print("reverse slicing :",arr[::-1])
print("step slicing :",arr[1:6:3])
print("negative index slicing :",arr[-4:])
```

```python
arr_2d = np.array([[2,3,4],
                   [0,9,8],
                   [12,4,5]])
print(arr_2d)
print("shape of the array:",arr_2d.shape)
print("size of the array:",arr_2d.size)
print("specific element",arr_2d[2,2])
print("printing row:",arr_2d[2])
print("printing column:",arr_2d[:,2])
```

## Sorting


```python
unsorted=np.array([4,3,1,6,3,2,6,9,232])

print("1d sorting:",np.sort(unsorted))

unsorted_2d=np.array([[4,3,1],[6,1,2],[1,9,2]])

print("2d sorting by column:",np.sort(unsorted_2d,axis=0))
print("2d sorting by row:",np.sort(unsorted_2d,axis=1))
```

## Filtering

```python
numbers = np.array([32,2,1,5,2,5,2,6])

even_numbers=[numbers % 2==0]
print(even_numbers)
```

## filter with mask

```python
mask=numbers>5
print("numbers greater than 5 :",mask,numbers[mask])
```

## Fancy indexing s np.where()

```python
indexing = [0,5,2]
print("using indexing:",numbers[indexing])

whereing = np.where(numbers>2)

print("usng where:",whereing,numbers[whereing])
```

```python
condition_array=np.where(arr_2d>2,arr_2d*3,"")
print(condition_array)
```

## Adding and removing in array

```python
a=np.array([1,2,3,4])
b=np.array([11,22,33,44,55])
c=np.array([11,21,31,41])

print(np.concatenate((a,b,c)))
```

## Array compatibility

```python
print("compatibily of arrays :",a.shape==b.shape==c.shape)
```

## Adding in 2d array in row wise and column wise

```python
print(arr_2d)

new_row=([11,22,33])

array_with_row=np.vstack((arr_2d,new_row))

print("array with new row: \n",array_with_row)

new_col=([0],[0],[0],[0])
array_with_col=np.hstack((array_with_row,new_col))
print("array with new column: \n",array_with_col)
```
